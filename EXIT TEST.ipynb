### Handling missing values

```python
df.head()
```

```python
for column in df.columns:
    if df[column].dtype=='float64':
        df[column].fillna(data[column].mean(),inplace=True)
```

```python
df.isna().sum()
```

```python
MD=df['mode'].mode()
df['mode'].fillna(value=MD,inplace=True)
```

```python
ND=df['voice_gender'].mode()
df['voice_gender'].fillna(value=ND,inplace=True)
```

```python
df.isna().sum()
```

```python
df=df.drop(['voice_gender'],axis=1)
```

```python
df=df.drop(['mode'],axis=1)
```

```python
df=df.drop(['musician_category'],axis=1)
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()


df['track_name']= label_encoder.fit_transform(df['track_name'])
df['key']= label_encoder.fit_transform(df['key'])
df['music_genre']= label_encoder.fit_transform(df['music_genre'])

df['key'].unique()
df['track_name'].unique()
df['music_genre'].unique()
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop('music_genre',axis=1)
```

```python
y=df['music_genre']
```

```python
y
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(x_train, y_train)
```

```python
y_predict = classifier.predict(x_test)
```

```python
y_predict
```

```python
print(confusion_matrix(y_test, y_predict))
print(classification_report(y_test, y_predict))
```
